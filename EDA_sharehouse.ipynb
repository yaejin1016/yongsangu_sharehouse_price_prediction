```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
share=pd.read_csv("share_final_data.csv",encoding="ANSI")
```

```python
share=pd.DataFrame(share);share.head()
```

```python
share.info()
```

```python
share.isnull().sum()
```

```python
share['인실']=share['인실'].astype(object)
share['건물타입']=share['건물타입'].astype(object)
share['방개수']=share['방개수'].astype(object)
share['화장실 개수']=share['화장실 개수'].astype(object)
share['난방방식']=share['난방방식'].astype(object)
share['층']=share['층'].astype(object)
share['총층']=share['총층'].astype(object)
share['승강기']=share["승강기"].astype(object)
share['주차']=share['주차'].astype(object)
share['편의점']=share['편의점'].astype(object)
```

```python
category_feature=[col for col in share.columns if share[col].dtypes == "object"]
category_feature
```

```python
numerical_feature=list(set(share.columns)-set(category_feature)-set(["adj_월세"]))
```

```python
numerical_feature
```

```python
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
```

# numerical_value 분포

```python
for col in numerical_feature:
    sns.distplot(share.loc[share[col].notnull(),col])
    plt.title(col)
    plt.show()
```

# Categorical_values 분포

```python
for col in category_feature:
    share[col].value_counts().plot(kind="bar")
    plt.title(col)
    plt.show()
```
